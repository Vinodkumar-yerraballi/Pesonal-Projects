```python
# Install the required libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import plotly.express as px
import plotly.io as pio
import matplotlib.image as mpimg
import plotly.graph_objects as go
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('C://Users//Admin//Downloads//archive//crimes_against_women_2001-2014.csv')
data.head()
```

# Data Preprocessing 

```python
#Remove the unwanted columns in the data set
data=data.drop(['Unnamed: 0'],axis=1)
data.info()
```

```python
#Checking the null values in the data set 
data.isna().sum()/len(data)*100
```

```python
#Unique States  in the data 
data['STATE/UT'].unique()
```

```python
#Covert the all the values into the upper cases
data['STATE/UT']=data['STATE/UT'].apply(lambda x: x.upper() if isinstance(x,str) else x)
data['STATE/UT'].unique()
```

```python
# create correlation map to understand the each columns correlation
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True,cmap='cubehelix',fmt='.2f',linewidths=1)
plt.show()
```

# Ask the Question From the Data
- To create a count plot depicting the number of values recorded from each state, we can use the available data to aggregate the counts and plot them. The count plot will provide a visual representation of the distribution of values across states.

- To understand the crime rate in a year-wise manner, we can create a line plot. The line plot will show the trend of the crime rate over the years, providing insights into any fluctuations or patterns in the data.

- Similarly, to analyze the number of recorded rape cases in each year, we can create another line chart. This line chart will allow us to observe the variations in rape cases over time, highlighting any significant changes or trends.

- To determine which state has the highest number of recorded rape cases, we can create a bar chart. The bar chart will compare the rape case counts across different states, making it easier to identify the state with the highest frequency of cases.

- Lastly, to identify the year with the highest number of recorded dowry cases, we can create a bar chart that includes state-wise data. This chart will help us visualize the distribution of dowry cases across different years and states, providing insights into the most affected regions.

```python
#Create countplot for states
plt.figure(figsize=(10,6))
sns.countplot(data=data,x=data['STATE/UT'])
plt.title("Visualize the States in the data")
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a line chart for year column
plt.figure(figsize=(10,7))
data['Year'].value_counts().plot(kind='line',color='red')
plt.title("Visualize the Years in the data")
plt.show()
```

```python
#Create a line chart to Visualize the sum of the rape cases in year wise
data.groupby('Year')['Rape'].sum()\
.plot(kind='line',color='black',figsize=(10,8))
plt.title("The Sum of the Rape cases with year wise")
plt.xlabel("Year")
plt.ylabel("Sum of the Rape cases")
plt.show()
```

```python
# According the 2001-14 which state have recorded more Rapes
data.groupby('STATE/UT')['Rape'].sum().sort_values(ascending=False)\
.plot(kind='bar',color=['#2ECCFA','#9AFE2E','#FA58D0','#2EFEC8','#F4FA58','#170B3B'],figsize=(12,8))
plt.title("How Many Rape cases recorded in Statewise")
plt.xlabel("Sates")
plt.ylabel("The Sum of the rape cases")
plt.show()
```

```python
# According to the 2001-14 Which Sates have Recorded Dowry Deaths case
data.groupby('STATE/UT')['Dowry Deaths'].sum().sort_values(ascending=False)\
.plot(kind='bar',color=['#0B2161','#A9F5F2','#CECEF6','#FFFF00','#FF00BF','#2EFE64'],figsize=(10,8))
plt.title("The Sum of the Dowry Death cases by Sate wise")
plt.xlabel("Sates")
plt.ylabel("The Sum of the Dowry Deaths Cases")
plt.show()
```

## Observations:
- From the data covering the years 2001 to 2014, we have discovered several interesting facts and trends.
- The countplot chart reveals that Uttar Pradesh and Madhya Pradesh have the highest number of recorded values compared to other states.
- The line chart depicting the crime rate over the years shows a consistent increase in crime rates year after year.
- The line chart representing the recorded rape cases clearly indicates a significant rise in cases each year, suggesting a concerning trend.
- According to the bar chart, Madhya Pradesh tops the list in terms of the highest number of recorded rape cases, followed by Bihar in the second position, indicating a higher incidence of these cases in these states.
- Analyzing the data on dowry cases, Uttar Pradesh emerges as the state with the highest number of recorded cases from 2001 to 2014, emphasizing the prevalence of this issue in the region.

```python
#How many rape Cases recorded in the AndhraPradesh State
# Create a Scatterplot and Under statnding the Rape cases with Year wise 
Andhra_pradesh=data[data['STATE/UT']=='ANDHRA PRADESH']
rape_cases_in_andhra=Andhra_pradesh.groupby('Year')['Rape'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,8))
sns.scatterplot(data=rape_cases_in_andhra,x=rape_cases_in_andhra.index,y=rape_cases_in_andhra.values,color='red')
plt.xlabel("Year")
plt.ylabel("Sum of the Rape Cases")
plt.title("Find the Rape cases recorded in Andhra Pradesh Sate")
plt.show()
```

```python
# Create a Scatter plot understanding the Dowry Death cases in Andhrapradesh
Dowry_death=Andhra_pradesh.groupby('Year')['Dowry Deaths'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,6))
sns.scatterplot(data=Dowry_death, x=Dowry_death.index, y=Dowry_death.values)
plt.show()
```

```python
#Create a line plot which each district in the Kidnapping and Abduction on year wise
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,8))
sns.lineplot(x="DISTRICT", y="Kidnapping and Abduction", hue="Year", alpha=.5, palette="tab10", data=Andhra_pradesh,linewidth=2.5)
plt.xticks(rotation=90)
plt.xlabel('DISTRICT')
plt.ylabel('Kidnapping and Abduction')
plt.show()
```

```python
# Create a bar chart undestanding   Kidnapping and Abduction recorded in 2001-14 year
kidnapping_abduction=Andhra_pradesh.groupby(['DISTRICT'])[['Kidnapping and Abduction']].sum()
kidnapping_abduction.sort_values(by='Kidnapping and Abduction',ascending=False).plot(kind='bar',figsize=(8,6),color=['red'])
plt.title("Visualize the Kidnapping and Abduction in the andhra pradesh state")
plt.xticks(rotation=90)
plt.xlabel('District')
plt.ylabel('Kidnapping and abduction')
plt.show()
```

# Another Some Intresting Question asked from the Data

```python
# Create Pie chart to undestand the Cruelty by Husband or His Relatives Cases with Year wise
Utterpradesh=data[data['STATE/UT']=='UTTAR PRADESH']
Utterpradesh.groupby('Year')['Cruelty by Husband or his Relatives'].sum().sort_values(ascending=False)\
.plot(kind='pie',title="Total Cruelty by Husband or his Relatives Cases Recorded in Uttar Pradesh",autopct='%1.2f%%',
    pctdistance=0.8,
    shadow=True,
    labeldistance=1.1)
```

```python
# Create a bar chart to visualize the top 10 district kidnapping and abdction cases in utterpradesh
new_data=Utterpradesh.where(Utterpradesh['Kidnapping and Abduction']>50)
new_data.groupby('DISTRICT')['Kidnapping and Abduction'].sum().sort_values(ascending=False).nlargest(10)\
.plot(kind='bar',title="Top 10 Districts in most kidnapping and abduction cases utterpradesh",figsize=(8,6),color=['#D2B48C','#CD853F','#BC8F8F','#DEB887'])
plt.xlabel("DISTRICT")
plt.ylabel("Sum of the most Kidnapping and abduction cases")
plt.show()
```

```python
# Create a separte data frame for Maharastra and find the some intresting facts.
Maharashtra=data[data['STATE/UT']=='MAHARASHTRA']
# Then we apply the where condition greate then 50 rape cases and finally visualize it
new_data_from_maharashtra=Maharashtra.where(Maharashtra['Rape']>50)
final_data=new_data_from_maharashtra.groupby('DISTRICT')['Rape'].sum().sort_values(ascending=False).nlargest(10)
#Visualize the above code in bar chart to undestand the insight from the data
fig=px.bar(final_data,x=final_data.index,y=final_data.values,title='Top District in Maharastra high Rape cases recorded',)
fig.show()
```

```python
# Calculate the all cases recorded against to the women in maharastra
total_cases=Maharashtra.groupby('Year').sum()
# Using the loop condition print the barplot
for i in total_cases:
    plt.figure(figsize=(10,6))
    sns.barplot(data=total_cases,x=total_cases.index,y=total_cases[i])
    plt.title(f'Total {i} Cases recorded in Maharastra against the womens')
    plt.xlabel("Year")
    plt.ylabel(f'Sum of the {i} cases recorded in Maharastra state')
    plt.show()
```

```python
finally_data=data.groupby(['STATE/UT']).mean()
finally_data=finally_data.drop('Year',axis=1)
for i in finally_data:
    plt.figure(figsize=(10,8))
    sns.lineplot(data=finally_data, palette="tab10", linewidth=2.5)
    plt.ylabel(f'Visualize the average {i} cases recorded in state wise')
    plt.xlabel("State")
    plt.xticks(rotation=90)
    plt.ylabel(f'Sum of the {i}  cases recorded in each state wise')
    plt.show()
```
