```python
#Let's visualize the top 30 countrys in the dataste
country=data['CGA'].value_counts().head(30)
#to visualize the data using seaborn
ax=plt.axes()
#set the background color
ax.set(facecolor='blue')
#set the figure size and style
sns.set(rc={'figure.figsize':(19,5)},style='darkgrid')
#Set the title and with fontsize 
ax.set_title("The top 30 country in the dataset",fontsize=30)
#using the bar plot visualize the country  in the dataset
sns.barplot(x=country.index,y=country,palette='rainbow')
#set country name on the x-axis
plt.xlabel("The countrys names")
#set counts name on the y-axis
plt.ylabel("# The value counts")
plt.xticks(rotation=60)
#let's visualise the data 
plt.show()
```

```python
#Let's find total gold medal in the dataset
gold=data['Gold'].sum()
#Find the total silver in the dataset
silver=data['Silver'].sum()
#Fing the total Bronze in the dataset
bronze=data['Bronze'].sum()
#Let's print the value
print("The total gold medal in the dataset",gold)
print("The total silver medal in the dataset",silver)
print("The total bronze medal in the dataset",bronze)
```

```python
plt.pie(x=[gold,silver,bronze],autopct='%1.2f%%',colors=['green','red','black'])
plt.title("Let's visualize the gold,silver and bronze percentage in the dataset")
plt.show()
```

```python
#To find the which countrys have the most medal in the dataset
medal=data.groupby(['CGA','Total']).size().reset_index().rename(columns={0:'Total_medals_country'}).sort_values(by='Total',ascending=False)
medal.head()
```

```python
#Lets' visualize the data using the plotly
fig=px.bar(medal[:30],x='CGA',y='Total',title='The top 30 country got the high medals in 2022 common wealth games',color_discrete_sequence=['red','yellow'])
fig.update_layout(bargap=0.5,plot_bgcolor='gray',template = "plotly_dark")
fig.show()
```

## How is The Distribution of Total Gold Medal Between Countries?

```python
plt.figure(figsize = (20,5))
plt.bar(data = data, x = 'CGA', height = 'Gold', width = 0.8, color = 'indigo', alpha = 0.6)
plt.xticks(fontsize = 8, rotation = 90, ha = 'center')
plt.ylim(0,80)
plt.xlabel('Country Name')
plt.ylabel('Total Gold Medal')
plt.title('The Distribution of Total Gold Medal Between Countries', fontsize = 14)
for i in range(len(data)):
    position = data['CGA'].loc[i]
    height = data['Gold'].loc[i] + 2
    text = str(data['Gold'].loc[i])
    plt.text(position,height,text,ha = 'center', fontsize = 8)
plt.show()
```

# SUMMARY
- If the **country has the lowest total silver medal or bronze medal**, it **does not mean** it **has the lowest total medal** in the standings
- It is **definitely** that if the **country has very high number of total gold medal**, it **does mean** that the **country may have high ranking** in the 2022 Common Wealth Games Medal Standings
- If **country has high number of total medal**, **country may have high ranking** in the 2022 Common Wealth Games Medal Standings.
